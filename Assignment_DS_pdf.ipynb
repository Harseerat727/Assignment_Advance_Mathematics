```python
import pandas as pd
import numpy as np

# Taking input
roll_number = 102317191

# load dataset
try:
    df = pd.read_csv("data.csv", encoding="utf-8")
except:
    df = pd.read_csv("data.csv", encoding="cp1252")

# Use NO2 column
x = df["no2"].dropna().values

# transform
ar = 0.05 * (roll_number % 7)
br = 0.3 * ((roll_number % 5) + 1)

z = x + ar * np.sin(br * x)

# estimating parameter
mu = np.mean(z)
sigma2 = np.var(z)

lam = 1 / (2 * sigma2)
c = 1 / np.sqrt(2 * np.pi * sigma2)

print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
